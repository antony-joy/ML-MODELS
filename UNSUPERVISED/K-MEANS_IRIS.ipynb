```python
# Import Dataset from sklearn
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
# Load Iris Data
iris = load_iris()
iris
```

```python
target_df = pd.DataFrame(data= iris.target, columns= ['species'])
print(target_df)

df = pd.DataFrame( iris.data, columns= iris.feature_names)
df
```

```python
df1=df[["petal length (cm)","petal width (cm)"]]
df1
```

```python
import matplotlib.pyplot as plt
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'])
plt.xlabel('length')
plt.ylabel('width')
```

```python
###############Preprocessing using min max scaler
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()

scaler.fit(df1[['petal length (cm)']])       
df1['petal length (cm)'] = scaler.transform(df1[['petal length (cm)']])

scaler.fit(df1[['petal width (cm)']])
df1['petal width (cm)'] = scaler.transform(df1[['petal width (cm)']])
```

```python
df1
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df1[['petal length (cm)','petal width (cm)']])
y_predicted
```

```python
df1['cluster']=y_predicted
df1
```

```python
km.cluster_centers_
```

```python
df2 = df1[df1.cluster==0]              ########for seperating the colours
df3 = df1[df1.cluster==1]
df4 = df1[df1.cluster==2]
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='beige')
plt.scatter(df4['petal length (cm)'],df4['petal width (cm)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('length')
plt.ylabel('width ($)')
plt.legend()


######it is not right cause of the age and income are miles away from each other
```

```python
################Elbow Plot
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df1[['petal length (cm)','petal width (cm)']])
    sse.append(km.inertia_)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
