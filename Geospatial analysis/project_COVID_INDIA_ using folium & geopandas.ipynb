## USING GEOCODING & MARKCLUSTER

```python
from geopandas.tools import geocode
from geopy.geocoders import Nominatim
import matplotlib
import folium
```

```python
import pandas as pd
import folium

covid_data=pd.read_csv("https://api.covid19india.org/csv/latest/state_wise.csv")
covid_data
```

```python
geolocator = Nominatim(user_agent="antonyjoy07@gmail.com")
location = geolocator.geocode("kerala")
location1=geolocator.geocode("Tamil Nadu")
print(location.address)
print(location1.address)
print((location.latitude, location.longitude))
print(location.point)
```

```python
lat,lng=[],[]
for i in covid_data.State:
    location = geolocator.geocode(i)
    lat.append(location.latitude)
    lng.append(location.longitude)

covid_data["Latitude"],covid_data["Longitude"]=lat,lng
covid_data
```

```python
covid=covid_data[["State","Active","Last_Updated_Time","Latitude","Longitude"]]

df=covid.drop([0,37])
df
```

```python
map=folium.Map((22.3511148, 78.6677428),zoom_start=4.4)

display(map)
```

```python
map=folium.Map((22.3511148, 78.6677428),zoom_start=4.4)
for lab, row in df.iterrows() :
    
    folium.Marker(location=(row['Latitude'],row['Longitude']),icon=folium.Icon(icon_color='red'),popup=row['State']+":"+str(row['Active']),tooltip=row['State']).add_to(map)
# Plotting 
map
```

```python
import numpy as np
label_ranges = [0, 1000, 20000, np.inf]
label_names = ['low', 'medium', 'high']

# Create new category column
df['category'] = pd.cut(df['Active'], bins = label_ranges, labels = label_names)
df

```

```python
map=folium.Map((22.3511148, 78.6677428),zoom_start=4.4)
for lab, row in df.iterrows() :
    if row["category"]=="low":
        colour_set='beige'
    elif row["category"]=="medium":
        colour_set='orange'
    else:
        colour_set='red'
    folium.Marker(location=(row['Latitude'],row['Longitude']),icon=folium.Icon(icon_color=colour_set),popup=row['State']+":"+str(row['Active']),tooltip=row['State']).add_to(map)
# Plotting 
map
```
