### REMOVING highly CORRELATED FEATURES &   RFE  &  PCA

ACCURACY=79 TO 90 %
PRECISION= MUCH BETTER THAN USING WITHOUT USING RFE... WE GET AVG ABOUT 85+

```python
#categorical data
categorical_cols = ["Sex","cp","fbs","restecg","exang","slope","ca","thal"]         #not skipping fbs

#import pandas as pd
df_dumm_nopresence_nonull = pd.get_dummies(df_nopresence_nonull, columns = categorical_cols)
df_dumm_nopresence_nonull
```

#### Remove highly correlated featuresÂ¶


```python
correlated_features = set()
correlation_matrix = df_dumm_nopresence_nonull.corr()


    
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)

print("HIGHLY CORRELATED FEATURES ARE",correlated_features) 
```

```python
df_rem_cor = df_dumm_nopresence_nonull.drop(df_dumm_nopresence_nonull[['Sex_1.0', 'restecg_2.0', 'fbs_1.0', 'slope_2.0', 'thal_7.0', 'exang_1.0']], axis=1)
df_rem_cor
```

#### Running RFECV

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier

X = df_rem_cor
target = dfnum_nonull.presence


rfc = RandomForestClassifier(random_state=101)
rfecv = RFECV(estimator=rfc, step=1, cv=StratifiedKFold(10), scoring='accuracy')
rfecv.fit(X, target)

print('Optimal number of features: {}'.format(rfecv.n_features_))      ###### they are saying that 22 features are good with .8 abs corr...
```

```python
plt.figure(figsize=(9, 9))
plt.title('Recursive Feature Elimination with Cross-Validation', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Number of features selected', fontsize=14, labelpad=20)
plt.ylabel('% Correct Classification', fontsize=14, labelpad=20)
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_, color='#303F9F', linewidth=3)
plt.show()
```

```python
print(np.where(rfecv.support_ == False)[0])                                 ######removing non-optimal features

X.drop(X.columns[np.where(rfecv.support_ == False)[0]], axis=1, inplace=True)      ##### nothing removed
```

```python
dset = pd.DataFrame()
dset['attr'] = X.columns
dset['importance'] = rfecv.estimator_.feature_importances_
dset = dset.sort_values(by='importance', ascending=False)

plt.figure(figsize=(9, 10))
plt.barh(y=dset['attr'], width=dset['importance'], color='#1976D2')
plt.title('RFECV - Feature importances', fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Importance', fontsize=14, labelpad=20)
plt.show()
```
