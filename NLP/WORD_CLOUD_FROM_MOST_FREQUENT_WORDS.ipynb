```python
import pandas as pd
import matplotlib.pyplot as plt
add1="C:/Users/ANTHONY/Desktop/Corona_NLP_test.csv"
corona=pd.read_csv(add1)
corona = corona.dropna(subset=['OriginalTweet']) 

corona
```

```python
corona_data = corona[corona.Location=="NYC"]
corona_data
```

```python
import nltk
nltk.download('stopwords')
```

```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

sw = set(stopwords.words("english"))
clou=""
most_occur=[]
for i in corona_data.OriginalTweet:
    text_tokens = word_tokenize(i)

    tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
    tokens_with_an = [word for word in tokens_without_sw if word.isalpha()]

    filtered_sentence = (" ").join(tokens_with_an)
    Coun = Counter(filtered_sentence.split())
    x=Coun.most_common()[0][0]
    clou = clou + " " +x
print(clou)
```

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt

# Generate the word cloud from the east_of_eden string
cloud_east_of_eden = WordCloud(background_color="white").generate(clou)

# Create a figure of the generated cloud
plt.imshow(cloud_east_of_eden, interpolation='bilinear')  
plt.axis('off')
# Display the figure
plt.show()
```
