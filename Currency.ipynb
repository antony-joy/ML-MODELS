```python
df1.columns = ['ds', 'y']

df1[2246:2248]
```

```python
df1.info()
```

```python
df1['y'] = pd.to_numeric(df1['y'], errors='coerce')
df1
```

```python
df1.dropna(inplace=True)
df1
```

```python
df1['ds'] =  pd.to_datetime(df1['ds'])
df1
```

```python
#some values cannot be converted to numeric cause its of different form ex- row 2247

#          ds         y
## 2247  2012-05-01   -

data_columns=['y'] 
df2 = (df1.drop(data_columns, axis=1) .join(df1[data_columns].apply(pd.to_numeric, errors='coerce')))

df2 = df2[df2[data_columns].notnull().all(axis=1)]

df2
```

```python
from fbprophet import Prophet
dir(Prophet)       
```

```python
model=Prophet()
df1.columns
```

```python
model.fit(df1)
```

```python
future_dates=model.make_future_dataframe(periods=600)             #########3  this shows future dates for 100 days 
future_dates
```

```python
prediction=model.predict(future_dates)                #########3 prediction or forecasting
prediction
```

```python
a=5724
s=6261
preddf=prediction.loc[5724:6261]
preddf=preddf[["ds","yhat"]]
preddf
```

```python
model.plot(prediction)
```

```python
model.plot_components(prediction)
```

```python
from fbprophet.plot import plot

plot(model, prediction, figsize=(20, 12))
```

```python
from fbprophet.diagnostics import cross_validation
df_cv = cross_validation(model, initial='1000 days', period='50 days', horizon = '100 days')
df_cv.head()
```

```python
from fbprophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.head()
```
