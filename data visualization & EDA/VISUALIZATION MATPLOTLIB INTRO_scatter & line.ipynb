### Choose Figure Style


```python
# Our style choices
plt.style.available
```

```python
# We are currently using default plot 
# so this will not change much about our plot
plt.style.use('fivethirtyeight')

plt.plot(df.Month,df ['Interest Paid'])
plt.plot(df.Month, df ['Principal Paid'])
```

## Setting plot titles, labels, and limits

```python
plt.plot(df.Month,df ['Interest Paid'],c= 'k')
plt.plot(df.Month, df ['Principal Paid'],c= 'b')

plt.xlim(left=1,right=70)
plt.ylim(bottom=0,top=1000)
```

```python

plt.style.use('classic')

plt.plot(df.Month,df ['Interest Paid'],c= 'k')
plt.plot(df.Month, df ['Principal Paid'],c= 'b')


plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid Each Month')
```

```python

plt.plot(df.Month,df ['Interest Paid'],c= 'r')
plt.plot(df.Month, df ['Principal Paid'],c= 'b')



plt.xlabel('Month', fontsize = 15)
plt.ylabel('Dollars', fontsize = 15)
plt.title('Interest and Principal Paid Each Month', fontsize = 15)
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 15)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
plt.plot(df.Month,df ['Interest Paid'],c= 'k')
plt.plot(df.Month, df ['Principal Paid'],c= 'b')


plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid Each Month')
```

### MARKER TYPE

```python
plt.plot(df.Month,df ['Interest Paid'],c = 'k', marker = '.', markersize = 10)
plt.plot(df.Month, df ['Principal Paid'],c = 'b', marker = '.', markersize = 10)

```

### GRID

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
plt.plot(df.Month,df ['Interest Paid'],c= 'k')
plt.plot(df.Month, df ['Principal Paid'],c= 'b')


plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid Each Month')

plt.grid()
```
